```python
def get_work_cols(df):
    return df[[col for col in df.columns if not 'META' in col and not 'PROCESS' in col]]

def prepare_X_y(df):
    dff = get_work_cols(df)
    min_prof = dff['profitability'].min()
    dff['profitability_moved'] = dff['profitability'] - min_prof + 1
    dff = dff.drop(['revenue','profitability'], axis=1)
    
    X = dff.drop('profitability_moved', axis=1)
    X.fillna(X.mean(), inplace=True)
    y = dff['profitability_moved']
    
    return X, y

def normalize_df(df):
    return (df - df.min())/(df.max() - df.min())
```

```python
df = pd.read_csv('15 topics.csv')
```

```python
list(df.columns)
```

```python
work_df = get_work_cols(df)
for col in list(work_df.columns):
    print(col)
    col_nan = len([x for x in work_df[col] if pd.isna(x)])
    print(f'nan: {col_nan}\t{col_nan/work_df[col].shape[0]}%\n')
```

```python
cast_cols = [c for c in list(work_df.columns) if 'crew__' in c]
crew_count = {}
for col in cast_cols:
    crew_count[col] = len([p for p in work_df[col] if pd.isna(p)])

sorted_crew = sorted(crew_count, key=lambda job: crew_count[job], reverse=True)
for job in sorted_crew:
    print(f'{crew_count[job]}\t{crew_count[job]/work_df.shape[0]}\t{job}')
    
```

```python
from sklearn.ensemble import ExtraTreesRegressor
```

```python
test_df = work_df.loc[(work_df.profitability < 1000) & (work_df.profitability > -1000)].reset_index(drop=True)
test_df = normalize_df(test_df)
X, y = prepare_X_y(test_df)

model = ExtraTreesRegressor()
model.fit(X, y)
```

```python
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(20).plot(kind='barh')
plt.show()
```
